```python
import csv
f = open("guns.csv", "r")
csvreader = csv.reader(f)
data = list(csvreader)
```

```python
print(data[:5])
```

```python
headers = data[:1]
data = data[1:]
print(headers)
print(data[:5])
```

```python
years = []
for row in data:
    years.append(row[1])
```

```python
years_count = {}
for year in years:
    if year in years_count:
        years_count[year] += 1
    else:
        years_count[year] = 1
        
print(years_count)
```

```python
import datetime
dates = [datetime.datetime(year = int(row[1]), month = int(row[2]), day = 1) for row in data]
```

```python
dates[:5]
```

```python
date_counts = {}

for date in dates:
    if date in date_counts:
        date_counts[date] += 1
    else:
        date_counts[date] = 1
        
```

```python
date_counts
```

```python
sexes = []
for row in data:
    sexes.append(row[5])
sex_counts = {}
for sex in sexes:
    if sex in sex_counts:
        sex_counts[sex] += 1
    else:
        sex_counts[sex] = 1
sex_counts
```

```python
races = [row[7] for row in data]
```

```python
race_counts = {}
for race in races:
    if race in race_counts:
        race_counts[race] += 1 
    else:
        race_counts[race] = 1
race_counts
```
